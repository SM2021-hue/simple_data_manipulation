```python
import pandas as pd
```

```python
myfile1 = pd.read_csv("file1.tsv",sep = "\t") 
print(myfile1.head(5))
```

```python
 myfile2 = pd.read_csv("file2.tsv",sep = "\t")  
print(myfile2.head(5))
```

```python
common_df = pd.merge(myfile1,myfile2,on = 'ENSGID') ##Q2 find common ENSGIDs,create df\n",
print(common_df)
```

```python
common_df['mean'] = common_df.mean(axis=1)
common_df['median'] = common_df.median(axis=1)
print(common_df)
```

```python
common_df['max'] = common_df.max(axis=1) 
```

```python
common_df['min'] = common_df[common_df > 0.000000].min(axis=1)
```

```python
common_df['Fold_Change'] = (common_df['max'] - common_df['min']) / common_df['min']
print(common_df)
```

```python
import numpy as np
```

```python
common_df['Log_Fold_Change'] = np.log(common_df['Fold_Change'])
print(common_df)
```

```python
df = common_df[['ENSGID', 'mean', 'median', 'Fold_Change', 'Log_Fold_Change']].copy()
print(df)
```
